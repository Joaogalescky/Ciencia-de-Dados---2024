```python
import pandas as pd
from tqdm.notebook import tqdm
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
df_tmp = pd.read_html("https://dadosabertos.ibama.gov.br/dados/SICAFI/AL/Arrecadacao/arrecadacaobenstutelados.html")
```

```python
bases = []
estados_brasileiros = ['AC', 'AL', 'AP', 'AM', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 'MT', 'MS', 'MG', 'PA', 'PB', 'PR', 'PE', 'PI', 'RJ', 'RN', 'RS', 'RO', 'RR', 'SC', 'SP', 'SE', 'TO']
```

```python
for estado in tqdm(estados_brasileiros):
  bases.append(pd.read_csv(f"https://dadosabertos.ibama.gov.br/dados/SICAFI/{estado}/Arrecadacao/arrecadacaobenstutelados.csv", sep=";"))
```

```python
base = pd.concat(bases)
base.sample(5)
```

```python
#Quantos CPF's tem na base?
cpfs = base['CPF/CNPJ'].unique().size
cpfs
```

```python
#Quantas multas tem?
multas = base['Nº AI'].unique().size
multas
```

```python
data = base['Data Auto'].str.split('/', expand=True)
data.columns = ['dia', 'mes', 'ano']
data.sample(2)
```

```python
base[['dia', 'mes', 'ano']] = base['Data Auto'].str.split('/', expand=True)
base.sample(2)
```

```python
#Total de valor das multas?
total_multa = base['Valor Pago'].unique().sum()
total_multa
```

```python
#Quantas multas em 2023 em Santa Catarina?
estado = 'SC'
ano = '2023'
base_estado = base[base['UF'] == estado]
base_ano = base_estado[base_estado['ano'] == ano]
base_ano['Nº AI'].unique().size
```

```python
#Quantas multas em 2023 em Corbelia/PR?
estado = 'PR'
ano = '2023'
municipio = 'Corbelia'
base_estado = base[base['UF'] == estado]
base_municipio = base_estado[base_estado['Município'] == municipio]
base_ano = base_estado[base_estado['ano'] == ano]
base_ano['Nº AI'].unique().size
```
