```python
import numpy as np
import pandas as pd
```

```python
# Biblioteca que oferece recursos de plotagem para a criação de gráficos 2D e 3D com visualizações estáticas, animadas e interativas
import matplotlib.pyplot as plt
```

```python
# -10 = inicio
# 10 = fim
# 40 = demarcações
val_x = np.linspace(-10,10,40)
val_x
```

```python
val_y = val_x**2
val_y
```

```python
plt.plot(val_x, val_y)
```

```python
plt.scatter(val_x, val_y)
```

```python
plt.scatter(val_x, val_y)
plt.plot(val_x, val_y)
```

```python
plt.scatter(val_x, val_y, color='k')
plt.plot(val_x, val_y, color='g')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de dispersão')
plt.show()
```

```python
import seaborn as sns # Biblioteca para fazer gráficos
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(x = 'sex', hue = 'smoker', data = tips)
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips)
# A 'barra' representa a variância, uma medida de dispersão que mostra o quão distante cada valor desse conjunto está do valor central (médio)
# Quanto menor é a variância, mais próximos os valores estão da média; mas quanto maior ela é, mais os valores estão distantes da média.
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips)
```

```python
# Diagrama de Candlestick
# Representa a abertura, fechamento, máximo e o mínimo de um título em um período determinado
# Consiste na representação gráfica do preço de um ativo ao longo de determinado período, podendo ser por 30 minutos, um dia ou um mês, por exemplo.
sns.boxplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips)
```
